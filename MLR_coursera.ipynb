#Multiple Linear Regression

#Importing Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import copy , math
np.set_printoptions(precision=2)  # reduced display precision on numpy arrays
```

```python
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

```python
print(f"Train {X_train}")
print(X_train.shape)
print(y_train.shape)
```

#Choosing W and b as per coursera values

```python
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
print(f"Shape of W parameter{w_init.shape},type of w parameter{type(w_init)}")
```

#Model Prediction With Multiple Variables

#Single Prediction element by element


```python
def single_prediction(x,w,b):
  n = x.shape[0]
  p = 0
  for i in range(n):
    p_i = w[i]*x[i]
    p = p+ p_i
  p = p + b
  return p
```

```python
# get a row from our training data
x_vec = X_train[0,:]
print(f"x_vec shape {x_vec.shape}, x_vec value: {x_vec}")
#making a predicition
f_wb = single_prediction(x_vec,w_init,b_init)
print(f"f_wb shape {f_wb.shape}, prediction: {f_wb}")
```

#using Vector , single


```python
def predict(x,w,b):
  p = np.dot(x,w) + b
  return p
```

```python
print(predict(x_vec,w_init,b_init))
```

#Cost Function

```python
def cost(X,y,w,b):
  m = X.shape[0]
  cost = 0.0
  for i in range(m):
    fwb = np.dot(X[i],w) + b
    cost = cost + (fwb - y[i])**2
  cost = cost / (2*m)
  return cost
```

```python
cost = cost(X_train,y_train,w_init,b_init)
print(f"Cost Function:{cost}")
```
