#Sigmoid Function

```python
def sigmoid(z):
 g =  1/(1+np.exp(-z))
 return g
```

```python
#Array on spaced numbers from -10 to 10
#Basically a list of possible z values, hum inhe kuch bhi le sakte hai (-infi,+infi)
z_tmp = np.arange(-10,11)
y = sigmoid(z_tmp)
# Code for pretty printing the two arrays next to each other
np.set_printoptions(precision=3)
print("Input (z), Output (sigmoid(z))")
print(np.c_[z_tmp, y])
```

#PLOT

```python
fig,ax = plt.subplots(1,1, figsize=(5,3))
ax.plot(z_tmp,y, c="r")
ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
plt.show()
```
